#Q5.Which actor has acted in the most number of movies? Deep dive into the movies, genres and profits corresponding to this actor.

```python
df.columns
```

```python
from ast import literal_eval
from collections import Counter

actor_counter = Counter()

for row in df['cast'].dropna():
  for actor in literal_eval(row):
    actor_counter[actor['name']] += 1

actor_counter.most_common(1)
most_actor = actor_counter.most_common(1)[0][0]
most_actor
```

```python
# extract all movies of this actor
actor_movies = df[
    df['cast'].apply(
        lambda x: pd.notna(x) and
        most_actor in [actor['name'] for actor in literal_eval(x)]
    )
]

actor_movies.head()
```

```python
actor_movies[['title','release_date','genres','budget','revenue','profit','roi']]
```

```python
actor_movies = actor_movies.copy()

actor_movies['genre_names'] = actor_movies['genres'].apply(
    lambda x: ', '.join(
        genre['name'] for genre in literal_eval(x)
    )
)

actor_movies[['title','genre_names','profit']]
```

```python
from collections import Counter

genre_counter = Counter()

for genres in actor_movies['genres']:
    genre_counter.update(
        genre['name'] for genre in literal_eval(genres)
    )

genre_df = (
    pd.DataFrame(
        genre_counter.items(),
        columns=['Genre','Movies']
    )
    .sort_values('Movies', ascending=False)
)

print(genre_df)
```

```python
actor_movies[['title','profit']].sort_values(
    by='profit',
    ascending=False
)
```

```python
best = actor_movies.loc[actor_movies['profit'].idxmax()]

print(best['title'])
print(best['profit'])
```

```python
worst = actor_movies.loc[actor_movies['profit'].idxmin()]

print(worst['title'])
print(worst['profit'])
```

```python
print("Total Movies :", len(actor_movies))
print("Average Profit :", actor_movies['profit'].mean())
print("Median Profit :", actor_movies['profit'].median())
print("Maximum Profit :", actor_movies['profit'].max())
print("Minimum Profit :", actor_movies['profit'].min())
```

```python
import matplotlib.pyplot as plt
color_List=['blue','red','green','yellow','orange','pink','purple','brown','gray','black','olive','silver']
genre_df.plot(
    x='Genre',
    y='Movies',
    kind='bar',
    color = color_List,
    figsize=(8,5)
)

plt.title(f"Genres of {most_actor}")
plt.ylabel("Number of Movies")
plt.show()
```
